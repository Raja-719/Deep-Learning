```python
import keras
from keras.applications.resnet50 import ResNet50
from keras.applications.resnet50 import preprocess_input, decode_predictions
from matplotlib.pyplot import imread, imshow
from tensorflow.keras.preprocessing import image
import numpy as np
import cv2

# Load the ResNet50 model
model = ResNet50(weights='imagenet')

# Load and preprocess the input image
img_path = 'Rose.jpeg'
img = keras.utils.load_img(img_path, target_size=(224, 224))
x = keras.utils.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# Predict the class of the input image
preds = model.predict(x)

# Display the top 3 predictions
print('Predicted:', decode_predictions(preds, top=3)[0])

# Load and preprocess the input image for visualization
img = cv2.imread(img_path)
img = cv2.resize(img, (224, 224))
x = np.expand_dims(img, axis=0)
x = preprocess_input(x)

# Display the input image
print('Input image shape:', x.shape)
my_image = imread(img_path)
imshow(my_image)

# Change every second color of the image
img_height, img_width, _ = img.shape
for i in range(0, img_height, 2):
    for j in range(0, img_width, 2):
        img[i, j] = [255, 0, 0]  # Change every second color to blue (you can modify this color)

# Display the modified image
imshow(img)
```
